# Creating and Plotting a Linear Equation and a Quadratic Equation with NumPy and Matplotlib

```python
x=np.linspace(start=-2,stop=2,num=100)
```

```python
#linear: y=m*x+c
m=2
c=3
y=m*x+c
```

```python
plt.plot(x,y,label='y=2x+3',color='green')
plt.title('Linear Equation')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.style.use("fast")
plt.xlim(-2,2)
plt.ylim(-1,7)
plt.legend()
plt.grid(True)
plt.show()
```

```python
#quardic equation:  ax^2+bx+c
a=2
x=np.linspace(start=-1,stop=7,num=200)
b=-4
c=4
```

```python
y=a*x**2+b*x+c
```

```python
plt.plot(x,y,label='2x^2+bx+c',color='black')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.style.use("grayscale")
plt.grid(True)
plt.legend()
plt.xlim(-5,10)
plt.ylim(0,30)
plt.show()
```

# Matplotlib Figure Object

```python
#create figure

fig=plt.figure()
# Add axes to the figure (left, bottom, width, height)
ax=fig.add_axes([0.8,0.8,0.5,0.5])
plt.show()
```

```python
x=np.linspace(0,10,100)
fi=plt.figure(figsize=(5,5))
ax=fi.add_axes([0.8,0.8,0.9,0.9])
y=np.cos(x)
ax.plot(x,y,label='cos(x)',color='green')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('Amplitude')
ax.set_title('plot of cos(x)')
plt.grid(True)


plt.show()
```

```python
#left, bottom,width,height
x=np.linspace(0,10,100)
aa=plt.figure(figsize=(8,5))
left=aa.add_axes([0.4,0.2,0.3,0.5])
left.plot(x,x+1,label='left')
left.set_title('left figure')
left.set_xlabel('x-axis')
left.set_ylabel('y-axis')
plt.grid(True)
bottom=aa.add_axes([0.2,0.8,0.3,0.5])
bottom.plot(x,x+2,label='bottom')
bottom.set_title('bottom figure')
bottom.set_xlabel('x-axis')
bottom.set_ylabel('y-axis')
plt.show()
```

# Matplotlib Subplots

```python
fig,ax=plt.subplots()
ax.plot([1,2,3,4],[1,8,9,16], label="x=x^3",color='orange')
plt.grid(True)
ax.legend()
plt.show()
```

```python
fig,ax=plt.subplots(2,2,figsize=(5,5),dpi=100, sharey=True, subplot_kw={'facecolor': 'lightgray'}) #dpi is the size of an image , 
#sharey means it will share the axis of y values
ax[0,0].plot([1,2,3,4],[1,8,27,64])
ax[0,1].plot([1,2,3,4],[1,4,9,16])
ax[1, 0].plot([1, 2, 3, 4, 5], [5, 4, 3, 2, 1])    
ax[1, 1].plot([1, 2, 3, 4, 5], [25, 16, 9, 4, 1])
#plt.style.use("ggplot")
#plt.grid(True)
plt.tight_layout()
plt.show()

```

# Matplotlib Plot Styles and Customizations

```python
x=[0,2,4,6,8]
y=[0,8,64,72,128]
plt.style.use("bmh")
plt.plot(x,y,label='x^3',linewidth=5,linestyle='--')
plt.plot([0,3,5,7,9],[0,27,125,256,80],label='line2',color='black',linestyle='dashdot',linewidth=8)
plt.legend(loc='upper right')
plt.show()
```

### Markers

Markers can be used to highlight individual data points on a line or scatter plot.
https://matplotlib.org/stable/api/markers_api.html
The **zorder** parameter determines the drawing order of plot elements.

```python
x=[0,2,4,6,8]
y=[0,8,64,72,128]
plt.style.use("default")
plt.plot(x,y,label='x^3',marker='o',markersize=8,markerfacecolor='green',zorder=1)
plt.plot([0,3,5,7,9],[0,27,125,256,80],label='line2',linestyle='--',color='black',marker='8',markersize=8,markerfacecolor='red')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

```python
# Plot with zorder
plt.plot([1, 2, 3, 4], [40, 30, 20, 10], color='red', linewidth=2, zorder=1,marker='o')  # Drawn first
plt.plot([1, 2, 3, 4], [10, 20, 30, 40], color='blue', linewidth=3,linestyle='--', zorder=2,marker='o',
         markersize=10,markerfacecolor='black')  # Drawn on top
plt.plot([1, 2, 3, 4], [10, 13, 15, 17], color='green', linewidth=4, zorder=3,marker='h',
         markersize=10,markerfacecolor='orange') # Drawn on top of blue
plt.xlabel('x-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Show plot
plt.show()
```

# Matplotlib Bar Graphs

```python
ca=['category A','category B','category C']
val=[10,5,15]

plt.bar(ca,val,edgecolor='red')
plt.style.use('grayscale')
plt.xlabel('CATEGORIES')
plt.ylabel('VALUES')
plt.title('BAR -GRAPH')

plt.show()
```

```python
# Data
categories = ['Category A', 'Category B', 'Category C']
values = [10, 15, 7]

# Creating horizontal bar graph
plt.barh(categories, values, color=['skyblue','green','red'],edgecolor='black')

# Adding title and labels
plt.title('Horizontal Bar Graph')
plt.xlabel('Values')
plt.ylabel('Categories')

# Show plot
plt.show()
```

```python
ca=['category A','category B','category C']
val1=[10,5,10]
val2=[5,2,4]
plt.bar(ca,val1,label='value1',color='orange')
plt.bar(ca,val2,label='value2',bottom =val2,color='red')
plt.legend()
plt.show()
```

```python
# Data
categories = ['Category A', 'Category B', 'Category C']
values1 = [10, 15, 7]
values2 = [5, 3, 2]
x = np.arange(len(categories))  # the label locations

# Width of the bars
width = 0.35

# Creating grouped bar graph
plt.bar(x - width/2, values1, width, label='Part 1',color='skyblue')
plt.bar(x + width/2, values2, width, label='Part 2')

# Adding title, labels, and legend
plt.title('Grouped Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')
#plt.xticks(x, categories)
plt.legend()

# Show plot
plt.show()
```

```python
# Data
categories = ['Category A', 'Category B', 'Category C']
values = [10, 15, 7]
errors = [1, 2, 1.5]

# Creating bar graph with error bars
plt.bar(categories, values, yerr=errors, capsize=2,color='brown')

# Adding title and labels
plt.title('Bar Graph with Error Bars')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show plot
plt.show()
```
