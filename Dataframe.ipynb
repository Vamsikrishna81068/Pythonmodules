# Grouping Data:
Grouping data is a powerful way to perform segment-wise analysis and break down the dataset into chunks based on some criteria.

```python
data=pd.read_excel("C:/Users/B.sanjay readdy/OneDrive/Desktop/EXCel/Copy of Sample - Superstore.xlsx")
```

```python
data
```

```python
data.head()
```

```python
reg=data.groupby('Region')
reg
```

```python
for Region,groupdata in reg:
    print(f"Genre: {Region}")
    print(groupdata)
    print()
    
    
```

```python
avg=data.groupby('Region')['Profit'].mean()
avg
```

```python
avg=data.groupby('Region')['Profit'].agg([np.mean,np.std,np.max,np.min])
avg
```

```python
data.count()
```

```python
data['Region'].value_counts()
```

```python
def re(ser):
    return ser.max()-ser.min()
data1=data.groupby('Region')['Profit'].agg(re)
data1
```
